This project analyzes production data from multiple wells stored in an Excel workbook. 

The goal is to consolidate data across wells, generate time-based production summaries, and visualize well performance for operational insights and reporting.

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
excel = pd.ExcelFile('KingFisher.xlsx')
sheets = excel.sheet_names

sheets
```

```python
tables = []

for sheet in sheets:
    df = pd.read_excel('KingFisher.xlsx', sheet)
    df['Well'] = sheet
    tables.append(df)

final_table = pd.concat(tables, ignore_index=True)

final_table.head()
```

```python
fig = px.line(
    final_table,
    x='DATE',
    y='Allocated Gas',
    color='Well',
    title='Gas Production by Well Over Time'
)

fig.show()
fig.write_image('../images/field_production_trend.png')
```

This plot compares production trends across wells, highlighting differences in performance and decline behaviour.

```python
final_table['DATE'] = pd.to_datetime(final_table['DATE'])
final_table.set_index('DATE', inplace=True)

processed_table = final_table.drop(columns=['FACNAME', 'FIELDNAME', 'Well'])

# Monthly and yearly summaries
monthly_data = processed_table.resample('ME').sum()
yearly_data = processed_table.resample('YE').sum()

monthly_data.head()
```

Resampling allows production data to be aggregated into monthly and yearly reports, supporting operational and financial analysis.

```python
for sheet in sheets:
    df = pd.read_excel('KingFisher.xlsx', sheet)

    fig, ax = plt.subplots()

    # Oil
    ax.plot(df['DATE'], df['Allocated Oil'], label='Oil Rate')
    ax.set_ylabel('Oil Rate [bbl/day]')

    # Gas
    ax_gas = ax.twinx()
    ax_gas.plot(df['DATE'], df['Allocated Gas'], label='Gas Rate')

    # Hours
    ax_hrs = ax.twinx()
    ax_hrs.plot(df['DATE'], df['Hrs Online'], label='Hours')
    ax_hrs.spines['right'].set_position(('axes', 1.2))

    plt.title(sheet)
    plt.savefig(f'../images/{sheet}.png', dpi=300)
    plt.close()
```

Multi-axis plots allow comparison of oil, gas, and operational hours, providing a more complete view of well performance.

- Production trends vary significantly across wells  
- Time-based aggregation supports reporting and forecasting  
- Multi-variable visualization improves understanding of well performance  
- Combining datasets enables field-wide analysis and decision-making  

This project demonstrates how multi-well production data can be consolidated and analyzed to support operational reporting and performance monitoring.

Such workflows are essential in oil and gas operations for tracking production and optimizing field performance.
